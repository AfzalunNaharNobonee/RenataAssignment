```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('/content/Assignment-for-Data-Scientist.csv')
```

```python
df
```

```python
df = df.dropna(axis=1, how='all')
```

```python
df = df.drop('ID', axis=1)
```

```python
df = df.dropna()
```

```python
print(df)
```

```python
categorical_columns = ['Division', 'Gender', 'MaritalStatus']
for column in categorical_columns:
    label_encoder = LabelEncoder()
    df[column] = label_encoder.fit_transform(df[column])
```

```python
df
```

```python
X = df
y = df['Division']
```

```python

division_data = df[['Division']]
kmeans = KMeans(n_clusters=8)
kmeans.fit(division_data)


df['Cluster Label'] = kmeans.labels_
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(data=df, x='Division', y='Age', hue='Cluster Label', palette='Set2')
plt.xlabel('Division')
plt.ylabel('Age')
plt.title('K-means Clustering based on Divisions')
plt.show()
```

```python
df.to_csv(r'/content/output_file.csv', index=False)
```
